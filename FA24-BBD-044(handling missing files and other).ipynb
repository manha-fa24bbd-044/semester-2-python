```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('melb_data.csv')
```

```python
#HOMEWORK TASK 1:: Handling Missing Values
#1.Identify columns with more than 20% missing values and drop them.
missing_person = df.isna().sum() / len(df) * 100
df_cleaned = missing_person[missing_person > 20].index
df_cleaned = df.dropna()
print(df_cleaned.info())

#2.Fill missing values in numerical columns with the median
if 'Number' in df.columns:
    df["Number"].fillna(df["Number"].median(), inplace=True)
    print("Filled missing values in numerical columns with median.")
else:
    print("Column 'Number' not found in DataFrame.")

#3.Fill missing values in categorical columns with the mode
if 'Category' in df.columns:
    df["Category"].fillna(3, inplace=True)
    print("Filled missing values in categorical columns with mode")
else:
    print("Column 'Category' not found in DataFrame.")

print(df.head())
```

```python
# HOMEWORK TASK 2:Advanced Filtering and Indexing

#1.Retrieving all properties located in ;Richmond; with a price greater than $1,000,000.
richmond_properties = df[(df['Suburb'] == 'Richmond') & (df['Price'] > 1000000)]
print(richmond_properties.head())

#2. Extracting only the Price, Suburb, and BuildingArea; columns for properties,
# where the land size is above 500 square meters.
df_filtered = df_cleaned.loc[df_cleaned['Landsize'] > 500, ['Price', 'Suburb']]
print(df_filtered.head())

#3.Finding the top 5 most expensive houses in the dataset using sorting.
df_sorted = df_cleaned.sort_values(by='Price', ascending=False)
print(df_sorted.head(5))
```

```python
#HOMEWORK TASK 3: Data Transformation & Feature Engineering
#1.Create a new column that calculates the price per room (Price / Rooms).
df_cleaned.loc[:,'Price_Per_Room'] = df_cleaned['Price'] / df_cleaned['Rooms']
print(df_cleaned.head())

#2.Convert the ;Date; column (if available) into a proper datetime format.
df_cleaned.loc[:,"Date"] = pd.to_datetime(df_cleaned["Date"], dayfirst=True)
print(df_cleaned.head())

#3.Extract the year of sale from the date and create a new column Year_Sold.
df_cleaned.loc[:,'Year_Sold'] = df_cleaned['Date'].dt.year
print(df_cleaned.head())
```

```python
#HOMEWORK TASK 4:Aggregation and Grouping
#1.Find the average price of properties in each suburb.
suburb_avg_price = df.groupby("Suburb")["Price"].mean()
print(suburb_avg_price)

#2.Find the total number of properties sold in each suburb.
properties_per_suburb = df.groupby("Suburb")['Price'].count()
print(properties_per_suburb)

#3.Identify the suburb with the highest average price.
highest_avg_price_suburb = suburb_avg_price.idxmax()
print(highest_avg_price_suburb)
```
